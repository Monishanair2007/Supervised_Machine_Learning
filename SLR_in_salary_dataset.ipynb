## SALARY PREDICTION DATASET


```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
#loading dataset
df =pd.read_csv('/content/SalaryPrediction.csv')
```

## Describing datasets
Experience Years → Numerical (Continuous)

Represents years of work experience.

Salary → Numerical (Continuous)

Represents salary in currency units.

# Preprocessing dataset

```python
# Features and target
X = df['Experience Years'].values.reshape(-1,1)
y = df['Salary'].values.reshape(-1,1)
y = y/1000
```

```python
#create a model
model = LinearRegression()
```

```python
#model training
model.fit(X , y)
```

```python
#apply prediction
y_predict = model.predict(X)
print(y_predict)
```

```python
#mean squared error
mean_squared_error(y, y_predict)
```

```python
#Root mean squared error
root_mean_squared_error = np.sqrt(mean_squared_error(y, y_predict))
print("RMSE Error:", root_mean_squared_error)
```

```python
#r2 score
r2 = r2_score(y , y_predict)
print("R2 Score:", r2)
print("R2 Error:", 1 - r2)
```

```python
#scatter plot
plt.scatter(X, y)
plt.plot(X, y_predict, color='red')
plt.xlabel('Experience years')
plt.ylabel('Salary')
plt.title('Simple Regression on Salary Dataset')
plt.show()
```

```python
#regression parameters
print("Intercept: ", model.intercept_[0])
print("Coefficient: ", model.coef_[0])
print(f"Equation : Y = {model.intercept_[0].round(2)} + {model.coef_[0][0].round(2)}X ")
```
